```python
import pyodbc
print(pyodbc.drivers())
```

```python
dados_conexao = (
    "Driver={PostgreSQL ANSI(x64)};"
    "Server=localhost;"
    "Database=PETILLO.OFC;"
    "UID=postgres;"
    "PWD=123456"
)

conexao = pyodbc.connect(dados_conexao)
print("Tudo certo")
```

```python
cursor = conexao.cursor()
```

```python
cursor.execute("SELECT table_name FROM information_schema.tables WHERE table_schema = 'wshop'")
tables = cursor.fetchall()
```

```python
print("Tabelas no esquema 'wshop':")
for table in tables:
    print(table[0])
```

```python
cursor.execute("SELECT column_name FROM information_schema.columns WHERE table_schema = 'wshop' AND table_name = 'pessoas'")
tables = cursor.fetchall()
```

```python
print("Colunas na tabela 'pessoas':")
for table in tables:
    print(table[0])
```

```python
cursor.execute("SELECT * FROM wshop.pessoas LIMIT 10")
rows = cursor.fetchall()
for row in rows:
    print(row)
```

```python
conexao.close()
print("Conexao encerrada")
```
